```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
```

```python
xt=torch.tensor([1,2,3,4,5],dtype=torch.float32)
yt=torch.tensor([2,4,6,8,10],dtype=torch.float32)
w=torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
b=torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

```

```python
class LinearRegression:
    def __init__(self, lr=0.01, n_epoch=10):
        self.lr=lr
        self.n_epoch=n_epoch
        self.weight=None
        self.bias=None

    def fit(self, X, y):

        self.weight=0
        self.bias=0

        for _ in range(self.n_epoch):
            dw = (2*(self.weight*X+self.bias-y)*X).mean()
            db= (2*(self.weight*X+self.bias-y)).mean()
    
            self.weight-=dw*self.lr
            self.bias-=db*self.lr
    
    def predict(self, X):
        return self.weight*X+self.bias
        
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred_line=model.predict(x)

fig=plt.figure()
plt.plot(x,y,'ro') 
plt.plot(x, y_pred_line)
plt.grid()
plt.show()
```

### Using torch

```python
class LR_T(nn.Module):
    def __init__(self, input, output, lr=0.01, n_epoch=10):
        super(LR_T, self).__init__()
        self.linear=nn.Linear(input, output)
        self.lr=lr
        self.n_epoch=n_epoch

    def __forward__(self, X):
        return self.linear(X)

        
    def fit(self, X, y):
        with torch.torch.inference_mode():
            self.linear.weight[0,0] = 0
            self.linear.bias[0] = 0
            
        lossfunc=nn.MSELoss()
        optimizer=torch.optim.SGD(self.linear.parameters(), lr=self.lr)  
        
        for _ in range(self.n_epoch):
            y_pred=self.__forward__(X)            
            loss=lossfunc(y, y_pred)    
            optimizer.zero_grad()
            loss.backward()   
            optimizer.step()            
            
    
    def predict(self, X):
        return self.__forward__(X)


        
```

```python
xt=torch.tensor([1,2,3,4,5],dtype=torch.float32).reshape(-1,1)
yt=torch.tensor([2,4,6,8,10],dtype=torch.float32).reshape(-1,1)
mymodel=LR_T(1,1)
mymodel.fit(xt,yt)
```

```python

with torch.inference_mode(): 
    y_pred_line=mymodel.predict(xt)
    fig=plt.figure()
    plt.plot(xt,yt,'ro') 
    plt.plot(xt, y_pred_line)
    plt.grid()
    plt.show()
```

Note 1: Sometimes we would put training outside the class. In that case, we would have just `__init__` and `__forward__` methods.

Note 2: The double underscore (in for example `__init__`) is just to tell us that this method is not designed to be called outside of the class.
