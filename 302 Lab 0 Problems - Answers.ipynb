### Problem 1

Let $f(x)=\frac{x-3}{\sqrt{x^2+1}}$. 

**a.** Find the limit of $f(x)$ at $x=0$.

```python
f=(x-3)/sp.sqrt(x**2+1)
L=sp.limit(f, x, 0, '+-')  
print("Limit at x=2 is ", L)
```

**b.** Plot $f(x)$ between $x=-20$ and $x=20$, marking a limit point from part **a.**. 

```python
fn=sp.lambdify(x,f,'numpy')

xval=np.linspace(-20,20,1000)

fig=plt.figure()

plt.plot(xval,fn(xval), color='blue')
plt.plot(0,L, marker="o", color='red')  

plt.grid()
plt.ylim(-5,5)
plt.title("Function and Limit")
plt.axhline(0, color='black', lw=2)  
plt.axvline(0, color='black', lw=2) 

plt.show()
```

### Problem 2

A study was done to compare the speed $v$ (in miles per hour) with the mileage $M$
(in miles per gallon) of an automobile. The results are shown in the table.


$$
\begin{aligned}
\begin{array}{|c|c|c|c|c|c|c|c|c|c|c|c|c|c|}
\hline
v & 15 & 20 & 25 & 30 & 35 & 40 & 45 & 50 & 55 & 60 & 65 & 70 & 75 \\
\hline
M & 22.3 & 25.5 & 27.5 & 29.0 & 28.8 & 30.0 & 29.9 & 30.2 & 30.4 & 28.8 & 27.4 & 25.3 & 23.3 \\
\hline
\end{array}
\end{aligned}
$$

**a.** Find the quadratic model for $y(v)$ and plot it.

```python
v=np.arange(15,80,5)
M=[22.3, 25.5, 27.5, 29, 28.8, 30, 29.9, 30.2, 30.4, 28.8, 27.4, 25.3, 23.3]

Model=polynomial.polyfit(v,M,2) 
Polymodel=Polynomial(Model)
Polymodel
```

```python
vval=np.linspace(0,90,1000)
mfit=polynomial.polyval(vval,Model)

fig=plt.figure()

plt.scatter(v,M, color='red')
plt.plot(vval,mfit, color= 'blue')

plt.title("Model for M(v)")
plt.xlabel('Velocity')
plt.ylabel('Mileage')
plt.grid()

plt.axhline(0, color='black', lw=2) 
plt.axvline(0, color='black', lw=2) 

plt.show()
```

**b.** At what velocity do we have the highest mileage?

```python
Mder=Polymodel.deriv()
Mder.roots()
```

**Answer:** The highest mileage happens when the velocity is about 45.5 m/h
