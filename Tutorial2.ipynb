### Indentation and Errors
Python uses indentation to denote things that are grouped together. To see this, I will use loops. There are two main types of loops: for-loop and while-loop. I will use for-loop in this example:

```python
Total=0              # we define a variable called Total and set its value to be 0.
for j in range(10):  # we start a for-loop. The paramenter j will go from 0 to 9 (note that j will not be 10)
    Total=Total+j    # we adding j to Total
Total                # We print the final value of Total
```

```python
Total=0              
for j in range(10):  
Total=Total+j        # No indentation, so it is not part of the loop
Total             
```

Notice we got an error. This is because the Python doesn't know what j is now. It is smart enough to see why the error happened in this case. Read the error and you will see it knows what happened. ### Arrays

Often we would like to create arrays of numbers. This is useful for plotting graphs for example. There are few ways to create arrays. 

First is `linspace`. For example, lets create an array that contains 100 equally spaced numbers between 0 and 1:

```python
np.linspace(0,1,100)
```

Now we will to the same, but using `arange`:

```python
np.arange(0,1,0.01)
```

Note the differences between these two commands. In linspace you specify amount of points you want to take "100", but in arange you specify the distance between points "0.01". Also, linspace includes the end point, while arange doesn't.

Finally, we can create an array by explicitly stating its elements using `array`.

```python
np.array([2,3,-1])
```

Finally, you can do computation using arrays. For example, suppose I have an array of numbers, and I want to evaluate square every number in the array. This can be done quite easily:

```python
L=np.arange(5)  # this will create an array from 0 to 5 of integers, not including 5.
S=L**2
S
```

We can also, select specific element of the array. For example, suppose I want to see third number in the array S defined above. S contains 0,1,4,9 and 16. So, third number is 4. Python as most programing languages starts counting from 0. So, first number is in the position 0, second number is the position 1, third number is iin the position 2 and so on. So we want the number in the position 2:

```python
S[2]
```

We can also count from the end. For example, if I want to get last number in the array, I can say S[-1], to get second to last I can say S[-2]:

```python
print(S[-1])
print(S[-2])
```

We can also pick sub-array from the array. For example, suppose we want fro second to fourth element (pay attention to indecies):

```python
S[1:4]
```

Here are few more examples. Try to understand the syntax in each example.

### Graphs and Plots
Next, let's look at plotting. We will plot the graph of $f(t)=\cos(2 \pi t)$. 

We will use pyplot library which is part of bigger matplotlib library.

We need to add new library. Usually, it is a better idea to add all libraries in the beginning of the notebook to keep track of everything, but for this tutorial I will not do that. 

```python
import matplotlib.pyplot as plt



t = np.arange(0, 5, 0.02)   
# This creates an array called t of values that go from 0 to 5 with a step 0.02. The graph is done point by point,
# so to make it look continuous and smooth, we just need to make step very small.
# You can also use linspace instead and take a lot of points.

fig=plt.figure()
# This creates a figure object that will coontain your figure
# You can also specify the size of your figure and other stuff if needed.

plt.plot(t, np.cos(2*np.pi*t), 'r--')   

# plot command creates a graph. Inside plot command you type an array of x-coordinates, then an array of y-coordinates,
# then other properties you want. In this example, t contains x-coordinates, y-coordinates are the values of the function.
#"r--" part is a color and type of the curve. Try others like b+ or g* or yo or just c

plt.show()
#display figure
```

Notice we will also use `figure` command. In the previous example, our graph was just a simple output. Now we will add an object called Figure and we can modify it. Usually, it is a good idea to include this command.

This piece of code is more intimidating. But most of this is just for making the graph pretty. While you dont need to memorize all of this, try to see what each line does.
```python
# x axis values
x = np.linspace(0,7,300)
# corresponding y axis values
y1 =np.sin(x)
y2 =np.cos(x)

fig=plt.figure(figsize=(8,6))
  
# plotting the points 
plt.plot(x, y1, color='red', linestyle='dashed', linewidth = 3, label='Sine curve')
plt.plot(x, y2, color='blue', linestyle='dotted', linewidth = 2, label='Cosine curve')  


# setting x and y axis range
plt.ylim(-2,2)
plt.xlim(0,7)
  
# naming the x axis
plt.xlabel('x - axis')
# naming the y axis
plt.ylabel('y - axis')
  
# giving a title to my graph
plt.title('Some cool customizations!', size=30)

# adds grid
plt.grid()

# outside color
fig.patch.set_facecolor('xkcd:mint green')

#show legend
plt.legend(facecolor='pink', shadow=True)

plt.show()
```

Another common way to create figures is to use `subplots` command. This command is quite useful as it allows you to plot multiple plots together.

```python

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

fig, ax = plt.subplots(1, 2, sharey=True)   
# fig is the figure, ax are axes of subplots. We can use ax to distinguish between multiple plots
# ax[0] would be the first plot and ax[1] is the second
# Next, inside we have "1,2", this means we will have 1 row and 2 column of subplots. So 2 plots total
# sharey=True means, the y-axis for both will be the same. This is useful property when you aranging plots side by side.

ax[0].plot(x, y)
ax[0].set_title('Sharing Y axis')
ax[1].plot(x, y,'--', color='red')

plt.show()
```

### Functions

If you have a piece of code that you need to execute in various parts of your notebook or you have a very specialized piece of code, it is often a good idea to have it separate as a function. For that we use `def`. This can also be used to create normal mathematical functions.

For example, let's create $f(x)=x^2+1$ and evaluate it at $x=4$.

```python
def f(x):                # f is the name of the function and x is an input
    return x**2+1        # `return` will return computed value
```

```python
f(4)
```

For simple functions we can also use `lambda` command to create function. We won't use it much, but it is something worth knowing.

```python
g=lambda x: x**2+1    # g is the name of the function.
```

```python
g(4)
```

However, if we want something more complicated we should just use `def`. 
Suppose we have an array of numbers and we want to find the largest number in it. Let's create a function that goes through the given list and outputs the largest value. Pay attention to indentations as well.

```python
def max (L):                 # max is the name of the function and L is a variable that denotes an input (which will be our array)
    maximum=L[0]             # we set initial value of maximum to be the first element of the array
    for i in range(len(L)):  # len(L) is the size of the array
        if L[i]>maximum:     # if new element is bigger then maximum, we redefine maximum
            maximum=L[i]
    return maximum           # this is the output of our function
```

Now lets find the maximum in different lists:

```python
A=np.arange(4)
B=np.arange(1,10,0.1)
C=np.sin(B)
```

```python
max(A)
```

```python
max(B)
```

```python
max(C)   
```
